```python
def get_i5_i7_from_fastq(fastq_path: str) -> tuple[list[str], list[str], list[str]]:
    f = gzip.open(fastq_path, "rb") if fastq_path.endswith(".gz") else open(fastq_path)
    i7_list, i5_list, ix_list = [], [], []
    for line in tqdm(f):
        line = line.decode() if fastq_path.endswith(".gz") else line
        if line.startswith("@M00312"):
            i7_i5 = line.strip().split(":")[-1]
            try:
                i7, i5 = i7_i5.split("+")
            except ValueError:
                ix_list.append(i7_i5)
                continue
            i7_list.append(i7)
            i5_list.append(i5)
    f.close()
    return i7_list, i5_list, ix_list
```
